```python
root = 'D:/desk/shanghai_date'
```

```python
#now generate the ShanghaiA's ground truth
part_A_train = os.path.join(root,'part_A_final/train_data','images')
part_A_test = os.path.join(root,'part_A_final/test_data','images')
part_B_train = os.path.join(root,'part_B_final/train_data','images')
part_B_test = os.path.join(root,'part_B_final/test_data','images')
path_sets = [part_A_test]
```

```python
img_paths = []
for path in path_sets:
    for img_path in glob.glob(os.path.join(path, '*.jpg')):
        img_paths.append(img_path)
```

```python
model = CSRNet()
```

```python
model = model.cuda()
```

```python
checkpoint = torch.load('0model_best.pth.tar')
```

```python
from matplotlib import cm as c

img = transform(Image.open('./Shanghai/part_A_final/test_data/images/IMG_100.jpg').convert('RGB')).cuda()


output = model(img.unsqueeze(0))

print("Predicted Count : ",int(output.detach().cpu().sum().numpy()))

temp = np.asarray(output.detach().cpu().reshape(output.detach().cpu().shape[2],output.detach().cpu().shape[3]))

plt.imshow(temp,cmap = c.jet)

plt.show()

temp = h5py.File('./Shanghai/part_A_final/test_data/ground_truth/IMG_100.h5', 'r')

temp_1 = np.asarray(temp['density'])

plt.imshow(temp_1,cmap = c.jet)

print("Original Count : ",int(np.sum(temp_1)) + 1)

plt.show()

print("Original Image")

plt.imshow(plt.imread('./Shanghai/part_A_final/test_data/images/IMG_100.jpg'))

plt.show()
```
